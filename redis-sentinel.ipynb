```python
SERVER_CERT = os.path.join(CERT_DIR, "redis.crt")
SERVER_KEY = os.path.join(CERT_DIR, "redis.key")
SERVER_CA_CERT = os.path.join(CERT_DIR, "redisCA.crt")
```

```python
from redis.sentinel import Sentinel 

 
sentinel = Sentinel(
    [('localhost', '26379'), ('localhost', '26380'), ('localhost', '26381')], 
    socket_timeout=10, 
    sentinel_kwargs={'ssl': True, 
    'ssl_certfile': SERVER_CERT,
    'ssl_keyfile': SERVER_KEY,
    'ssl_ca_certs': SERVER_CA_CERT},
)

master = sentinel.master_for(
    'redismaster', 
    socket_timeout=10,
    ssl=True, 
    ssl_certfile=SERVER_CERT,
    ssl_keyfile=SERVER_KEY,
    ssl_ca_certs=SERVER_CA_CERT
)
```

```python
sentinel.discover_master('redismaster') 
```
